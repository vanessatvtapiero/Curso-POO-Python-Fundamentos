# üß™ Ejercicio 1: Registro de Empleados
## Enunciado:

* Crea una clase Empleado que contenga:

* Atributo p√∫blico: nombre

* Atributo protegido: _salario

* Atributo privado: __id

* Un contador de empleados como atributo de clase

* M√©todo m√°gico __str__ para imprimir la informaci√≥n del empleado

* M√©todo est√°tico es_valido_id que retorne True si un ID comienza con "EMP"

* Un m√©todo de clase que devuelva cu√°ntos empleados se han creado
* Tambi√©n usa @property y @setter para validar que el salario no sea menor de 1,000,000.

```python
class Empleado:
    contador_Empleados = 0
    def __init__(self, nombre, salario, ID):
        self.nombre = nombre
        self._salario = salario
        self.__ID = ID 
        Empleado.contador_Empleados += 1
    @property
    def salario(self):
        return self._salario
    @salario.setter
    def salario(self, valor):
        if valor < 1000000:
            raise ValueError("El salario no puede ser menor a 1,000,000")
        self._salario = valor

    def __str__(self):
        return f"Nombre: {self.nombre}, Salario: {self._salario}, ID: {self.__ID}"
    @staticmethod
    def es_valido_id(ID):
        return ID.startswith("EMP")
    @classmethod
    def obtener_contador_empleados(cls):
        return cls.contador_Empleados
Empleado1 = Empleado("Juan", 1500000, "EMP001")
Empleado2 = Empleado("Ana G√≥mez", 2000000, "EMP002")
print(Empleado1)
print(Empleado2)

print(f"Total de empleados: {Empleado.obtener_contador_empleados()}")

print(Empleado.es_valido_id("EMP003"))
print(Empleado.es_valido_id("12345"))
```

# üß™ Ejercicio 2: Cuenta Bancaria
## Enunciado:

* Crea una clase CuentaBancaria que incluya:

* Atributo p√∫blico: titular

* Atributo protegido: _saldo

* Atributo privado: __numero_cuenta

* M√©todo m√°gico __repr__ para representar la cuenta

* M√©todo depositar y retirar

* @property y @setter para acceder/modificar el saldo (no permitir saldo negativo)

* M√©todo de clase desde_string que cree una cuenta desde una cadena con formato: "Luis,1234,500000"

* M√©todo est√°tico es_cuenta_valida que verifique que el n√∫mero de cuenta tenga 4 d√≠gitos

```python
class CuentaBancaria:
    #Constructor de la clase CuentaBancaria
    def __init__(self, titular, saldo, nueva_cuenta):
        self.titular = titular
        self._saldo = saldo #protegido
        self.__nueva_cuenta = nueva_cuenta #PRIVADO

    def __repr__(self): #Devuelve un representacion en cadena de la instancia 
        return f"CuentaBancaria(titular={self.titular}, saldo={self._saldo}, nueva_cuenta={self.__nueva_cuenta})"
    
    def depositar(self, cantidad):
        if cantidad <=0: # NOS PERMITE AGREGAR VALOR A LA CANTIDAD QUE SE DESEA DEPOSITAR Y SE ESTA VALIDANDO DE QUE SEA MAYOR A 0 Y SI NO NOS MUESTRA EL ERROR Y SE SUMA AL SALDO 
            raise ValueError("La cantidad  adepositada debe ser mayor a 0 ")
        self.saldo += cantidad
        print(f"Dep√≥sito realizado. Nuevo saldo: {self._saldo}")

    def retirar(self, cantidad): #nOS PERIMTE RETIRAR EL SALDO DE LA CUENTA Y DESPUES VALIDAD  DE QUE LA CANTIDAD SEA MAYOR QUE 0 Y QUE HALLA SUFICIENTE SALDO PARRA RETIRAR Y SI SE CUMPLE SE RESTA LA CANTIDAD DEL SALDO
        if cantidad <= 0:
            raise ValueError("La cantidad a retirar debe ser mayor a 0 ")
        if cantidad > self._saldo:
            raise ValueError("No hay suficiente saldo para realizar la operaci√≥n")
        self._saldo -= cantidad
        print(f"Retiro realizado. Nuevo saldo: {self._saldo}")

    @property
    def saldo(self): #Este hace que devuelva el saldo actual 
        return self._saldo

    @saldo.setter #En este nos permite modificar el saldo y despues asegurando que no sea negativo
    def saldo(self, valor):
        if valor < 0:
            raise ValueError("El saldo no puede ser negativo")
        self._saldo = valor

    @classmethod # tiene acesso a los atributos y metodos de la clase pero no a los atributos de isntancia.
    def desde_string(cls, datos_cuenta):
        titular, saldo, nueva_cuenta = datos_cuenta.split(",") #Cre una instancia para partir de una cadena de formato 
        saldo = float(saldo)
        if not cls.es_valido_nueva_cuenta(nueva_cuenta): #Se extraen valores para que se validen el numero de cuenta y crea una instancia de la clase 
            raise ValueError("el numero de cuenta debe ser 4 d√≠gitos") # vERIFICA QUE EL NUMERO DE CUENTA TENGA MAXIMO DE 4 DIGITOS
        return cls(titular, saldo, nueva_cuenta)
    
    @staticmethod #Este metodo no tiene acceso a los atributos de la clase ni a los de instancia, solo a los parametros que recibe
    def es_valido_nueva_cuenta(nueva_cuenta):
        return len(nueva_cuenta) == 4 and nueva_cuenta.isdigit()
    #EJEMPLOS DE USO DE LA CLASE CuentaBancaria
cuenta1 = CuentaBancaria("Juan P√©rez", 500000, "1234")
print(cuenta1)
cuenta1.depositar(200000)
cuenta1.retirar(100000)
print(f"Saldo actual: {cuenta1.saldo}")
# Crear una cuenta desde una cadena
cuenta2 = CuentaBancaria.desde_string("Ana G√≥mez,1000000,5678")
print(cuenta2)

# Verificar la validez de un n√∫mero de cuenta
print(CuentaBancaria.es_valido_nueva_cuenta("1234"))
print(CuentaBancaria.es_valido_nueva_cuenta("12345"))
print(CuentaBancaria.es_valido_nueva_cuenta("12a4"))

```

# üß™ Ejercicio 3: Producto con Descuento
## Enunciado:

* Crea una clase Producto que tenga:

* Atributo p√∫blico: nombre

* Atributo protegido: _precio

* Atributo privado: __codigo

* M√©todo m√°gico __eq__ para comparar productos por su c√≥digo

* M√©todo est√°tico aplicar_descuento(precio, porcentaje) que calcule el nuevo precio

* Atributo de clase impuesto que todos los productos comparten

* @property y @setter para acceder y modificar el precio (no debe ser menor de 0)

```python
class Producto:
    impuesto = 0.21
    def __init__(self, nombre, precio, codigo):
        self.nombre = nombre
        self._precio = precio
        self.__codigo = codigo
    def __eq__(self, otro): # este metodo es para comparar si dos objetos son iguales.
        if isinstance(otro, Producto):
            return  self.__codigo == otro.__codigo
        return False
    @staticmethod
    def aplicar_descuento(precio, porcentaje):
        if porcentaje < 0 or porcentaje > 100:
            raise ValueError("El descuento debe estar entre 0 y 100")
        descuento = precio * (porcentaje / 100)
        nueva_precio = precio - descuento
        return nueva_precio
    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, valor):
        if valor < 0:
            raise ValueError("El precio no puede ser menor de 0.")
        self._precio = valor
    def __repr__(self):
        return f"Producto(nombre={self.nombre}, precio={self._precio}, codigo={self.__codigo})"
producto1 = Producto("Laptop", 1500000, "LAP001")
producto2 = Producto("Celular", 800000, "CEL001")

print(producto1)
print(producto2)

# Comparar productos por c√≥digo
print(producto1 == producto2)  # False

#aplicar descuento 
nuevo_precio = Producto.aplicar_descuento(producto1._precio, 10)
print(f"Nuevo precio despu√©s del descuento: {nuevo_precio}")
# Modificar el precio
producto1.precio= 1200
print(f"Precio actualizado del producto: {producto1._precio}")
# Intentar establecer un precio negativo (esto lanzar√° una excepci√≥n)
try:
    producto1.precio = -500
except ValueError as e:
    print(f"Error: {e}")  # Salida: El precio no puede ser menor de 0.



    
```

# üßÆ Ejercicio: Clases de Figuras Geom√©tricas
## üéØ Enunciado
### Crea una clase base llamada Figura con las siguientes caracter√≠sticas:

‚úÖ Requisitos:
* Atributos:
* P√∫blico: nombre

* Protegido: _color

* Privado: __id_figura (√∫nico por figura)

### M√©todos:
* M√©todo m√°gico __str__ para mostrar la figura

* M√©todo de clase crear_con_nombre que permite crear figuras por nombre (Figura.crear_con_nombre("Cuadrado"))

* M√©todo est√°tico es_color_valido(color) que retorna True si el color est√° entre ["rojo", "azul", "verde"]

* @property y @setter para acceder y modificar el color (solo si es v√°lido)

### Luego crea una subclase Circulo que herede de Figura:

‚úÖ Requisitos:
### Atributos:
* Protegido: _radio

### M√©todos:
* M√©todo m√°gico __eq__ que compara dos c√≠rculos por radio

* M√©todo area() que devuelva el √°rea del c√≠rculo

* @property y @setter para acceder/modificar el radio (debe ser mayor que 0)

```python
class Figura:
    def __init__(self, nombre, color, id_figura):
        self.nombre = nombre
        self._color = color
        self.__id_figura = id_figura

    def __str__(self):
        return f"Figura: {self.nombre}, Color: {self._color}, ID: {self.__id_figura}"

    @classmethod
    def crear_con_nombre(cls, nombre):
        return cls(nombre, "rojo", id_figura=id(nombre))  # Asignar un color por defecto y un ID √∫nico

    @staticmethod
    def es_color_valido(color):
        return color in ["rojo", "azul", "verde"]

    @property
    def color(self):
        return self._color

    @color.setter
    def color(self, valor):
        if not self.es_color_valido(valor):
            raise ValueError("Color no v√°lido. Debe ser 'rojo', 'azul' o 'verde'.")
        self._color = valor

import math

class Circulo(Figura):
    def __init__(self, nombre, color, id_figura, radio):
        super().__init__(nombre, color, id_figura)
        self._radio = radio

    def __eq__(self, otro):
        if isinstance(otro, Circulo):
            return self._radio == otro._radio
        return False

    def area(self):
        return math.pi * (self._radio ** 2)

    @property
    def radio(self):
        return self._radio

    @radio.setter
    def radio(self, valor):
        if valor <= 0:
            raise ValueError("El radio debe ser mayor que 0.")
        self._radio = valor

    def __str__(self):
        return f"{super().__str__()}, Radio: {self._radio}, √Årea: {self.area():.2f}"

# Crear una figura
figura1 = Figura.crear_con_nombre("Cuadrado")
print(figura1)

# Validar color
print(Figura.es_color_valido("azul"))  # Salida: True
print(Figura.es_color_valido("amarillo"))  # Salida: False

# Crear un c√≠rculo
circulo1 = Circulo("C√≠rculo 1", "azul", id_figura=id("C√≠rculo 1"), radio=5)
print(circulo1)

# Comparar c√≠rculos
circulo2 = Circulo("C√≠rculo 2", "verde", id_figura=id("C√≠rculo 2"), radio=5)
print(circulo1 == circulo2)  # Salida: True

# Modificar el radio
circulo1.radio = 7
print(circulo1)

# Intentar establecer un radio inv√°lido
try:
    circulo1.radio = -3
except ValueError as e:
    print(e)  # Salida: El radio debe ser mayor que 0.
```
